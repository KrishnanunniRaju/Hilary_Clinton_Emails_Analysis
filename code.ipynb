Data Preprocessing Steps
 1. Remove Empty Bodies
 2. Remove Stop words and punctuations
 3. Only select nouns, adjectives, verbs and adverbs

```python
data_preproccessed=data[pd.notnull(data['ExtractedBodyText'])]
data_preproccessed.shape
```

```python
from nltk import RegexpTokenizer
from nltk.corpus import stopwords
import string
stop_words = set(stopwords.words('english')+list(string.punctuation))
```

```python
tokenizer = RegexpTokenizer(r'\w+')
texts = [tokenizer.tokenize(email.lower()) for email in data_preproccessed['ExtractedBodyText']]
```

```python
import spacy
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

```python
def delete_stopwords(tokenized_sentence: list):
    return list(filter(lambda x: x not in stop_words, tokenized_sentence))
```

```python
texts=lemmatization(texts, allowed_postags=['NOUN', 'VERB', 'ADV'])
texts = list(filter(lambda x: len(x) > 5, [delete_stopwords(text) for text in texts]))
```

```python
from gensim import corpora
corpora_dict = corpora.Dictionary(texts)
corpora_dict.filter_extremes(no_below=15, no_above=0.5, keep_n=100000)
corpus = [corpora_dict.doc2bow(text) for text in texts]
```

# LDA MODEL

```python
from gensim import matutils
from gensim.models.ldamodel import LdaModel
```

```python
model_lda = LdaModel(corpus, passes=20, num_topics=10, id2word=corpora_dict)
```

```python
from gsdmm import MovieGroupProcess
```

```python
gsdmm_model=MovieGroupProcess(K=15, alpha=0.1, beta=0.3, n_iters=15)
y = gsdmm_model.fit(texts, len(corpora_dict))
```

```python
#top words per topic
def top_words(cluster_word_distribution, top_cluster, values):
    for cluster in top_cluster:
        sort_dicts = sorted(cluster_word_distribution[cluster].items(), key=lambda k: k[1], reverse=True)[:values]
        print("\nCluster %s : %s"%(cluster, sort_dicts))
```

```python
doc_count = np.array(gsdmm_model.cluster_doc_count)
print('Number of documents per topic :', doc_count)
# Topics sorted by the number of document they are allocated to
top_index = doc_count.argsort()[-15:][::-1]
print('Most important clusters (by number of docs inside):', top_index)

# get top words in topics
top_words(gsdmm_model.cluster_word_distribution, top_index, 20)
```
